# Create Data set

```python
id = {101,102,103}
print(type(id))
print(id)
for i in id:
    print(i)


a = {1,2,3,4,4,5,3,6,6,5,4,1,2,2}
print("Value is a :",a)
```

# Data set add data 

```python
numbers = {11,3,45,56,3,666,48,998}
print(numbers)
numbers.add(32)
print(numbers)
```

# Data set remove data

```python
numbers = {11,3,45,56,3,666,48,998}
print(numbers)
numbers.remove(666)
print(numbers)
```

# Data set Union

```python
a = {1,2,3,44,556,55,}
b = {4,55,77,55,88,99}
print("Union value is : ",a|b)
```

# Data set Intersection

```python
a = {1,2,3,44,556,55,4,99,88}
b = {4,55,77,55,88,99}
print("Intersection value is : ",a&b)
```

# Update Python Set

```python
companies = {'Lacoste', 'Ralph Lauren'}
tech_companies = ['apple', 'google', 'apple']

companies.update(tech_companies)

print(companies)

# Output: {'google', 'apple', 'Lacoste', 'Ralph Lauren'}
```
